```python
#Reading data into a dataframe
titanic = pd.read_csv('/content/drive/MyDrive/KDD/Titanic.csv')
```

```python
titanic.head(10)
```

```python
# Map the 'Survived' variable to Boolean values
titanic['Survived'] = titanic['Survived'].map({'Yes': 1, 'No': 0})
```

```python
#Transform into Pandas DataFrame
# One-hot encode categorical variables: 'Class', 'Sex', and 'Age'
categorical_columns = ['Class', 'Sex', 'Age']
encoder = OneHotEncoder(sparse_output=False, drop=None)
encoded_columns = encoder.fit_transform(titanic[categorical_columns])
encoded_df = pd.DataFrame(encoded_columns, columns=encoder.get_feature_names_out(categorical_columns))
titanic = titanic.drop(categorical_columns, axis=1)
titanic = pd.concat([titanic, encoded_df], axis=1)
```

```python
titanic
```

```python
#Asked 35 frequent itemsets with the support level 0.1

from mlxtend.frequent_patterns import fpgrowth, association_rules
frequent_itemsets = fpgrowth(titanic, min_support=0.1, use_colnames=True)
```

```python
print(frequent_itemsets)
```

```python
top_35_itemsets = frequent_itemsets.head(35)
```

```python
#Correct initial result of the associate rules with the confidence level 0.7
association_rules_df = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.7)
association_rules_df
```

```python
#consequents only being a value of the Survived variable
association_rules_df = association_rules_df[association_rules_df['consequents'].apply(lambda x: len(x) == 1 and 'Survived' in x)]
association_rules_df
```

```python
# Now display valid association rules with the relevant information
association_rules_df[['antecedents', 'consequents', 'support', 'confidence']]
```

Summary of Titanic Survival Analysis using Association Rule Mining

This notebook aimed to discover interesting relationships between passenger characteristics and survival on the Titanic using association rule mining.

The process involved:

Data Loading and Preparation: The Titanic dataset was loaded and preprocessed. This included converting the 'Survived' column to numerical values (1 for Yes, 0 for No) and applying one-hot encoding to the categorical features ('Class', 'Sex', and 'Age') to prepare the data for the association rule mining algorithm.

Frequent Itemset Generation: The FPGrowth algorithm was used to identify frequent itemsets (combinations of characteristics) that appear together in the dataset with a minimum support of 10%.
Association Rule Mining: Association rules were generated from the frequent itemsets with a minimum confidence threshold of 70%. These rules highlight potential relationships between different characteristics and survival.

Analysis of Survival Rules: The generated rules were filtered to focus specifically on those where 'Survived' was the consequent. This allowed for the identification of characteristics or combinations of characteristics that strongly predict survival.
Key Findings (Based on the filtered rules with Survived as consequent):

The analysis revealed strong associations between certain passenger characteristics and survival:

The rule {Sex_Female} -> {Survived} shows that being female is strongly associated with survival, with a confidence of approximately 73.2%.
The rule {Sex_Female, Age_Adult} -> {Survived} indicates that being an adult female is also strongly associated with survival, with a confidence of approximately 74.4%.
These findings suggest that gender, and to some extent age, played a significant role in survival rates on the Titanic. The higher confidence values for rules involving 'Sex_Female' highlight the well-known "women and children first" protocol that was followed during the disaster.
