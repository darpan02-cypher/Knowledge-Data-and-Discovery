```python
dataset = pd.read_csv("/content/drive/MyDrive/KDD/Market_Basket_Optimisation.csv", header = None)
dataset.shape
```

```python
dataset
```

```python
# Gather All Items of Each Transactions into Numpy Array
items = []
for i in range(0, dataset.shape[0]):
    for j in range(0, dataset.shape[1]):
        items.append(dataset.values[i,j])
# converting to numpy array
items = np.array(items)
print(items)
```

```python
#  Transform Them a Pandas DataFrame
df = pd.DataFrame(items, columns=["items"])
# Put 1 to Each Item For Making Countable Table, to be able to perform Group By
df["incident_count"] = 1
```

```python
df
```

```python
#  Delete NaN Items from Dataset
indexNames = df[df['items'] == "nan" ].index
df.drop(indexNames , inplace=True)
# Making a New Appropriate Pandas DataFrame for Visualizations
df2 = df.groupby("items").sum().sort_values("incident_count", ascending=False).reset_index()
```

```python
df2
```

```python
# Transform Every Transaction to Seperate List & Gather Them into Numpy Array
transactions = []
for i in range(dataset.shape[0]):
    transactions.append([str(dataset.values[i,j]) for j in range(dataset.shape[1])])
```

```python
#transaction is a nested list
transactions
```

```python
type(transactions)
```

```python
transactions_array = np.array(transactions)
```

```python
transactions_array
```

```python
transactions_array.shape
```

```python
# importing the required module
from mlxtend.preprocessing import TransactionEncoder
# initializing the transactionEncoder
te = TransactionEncoder()
te_trans = te.fit(transactions_array).transform(transactions_array)
```

```python
dataset2 = pd.DataFrame(te_trans, columns=te.columns_)
# dataset after encoded
dataset2
```

```python
# select top 30 items
first30 = df2["items"].head(30).values
# Extract Top 30
dataset3 = dataset2.loc[:,first30]
# shape of the dataset
dataset3.shape
```

```python
dataset3
```

```python
#Importing Libraries
from mlxtend.frequent_patterns import fpgrowth
#running the fpgrowth algorithm
#use_colnames=True: This ensures that the resulting frequent itemsets are displayed using the column names from the input dataset3, rather than the internal column indices
itemsets=fpgrowth(dataset3,min_support=0.05, use_colnames=True)
itemsets
```

```python
# importing required module
from mlxtend.frequent_patterns import association_rules
# creating asssociation rules
rules=association_rules(itemsets, metric="lift", num_itemsets=1)
# printing association rules
rules
```

```python
# Sort values based on lift
rules.sort_values("lift",ascending=False)
```
